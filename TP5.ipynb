```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:,0], X[:,1], color='#6667AB')
```

Même si nous connaissons déjà le nombre optimal de clusters, on va lui déterminer à l'aide de la méthode du coude (elbow method). Pour obtenir les valeurs utilisées dans le graphique, nous entraînons plusieurs modèles en utilisant un nombre différent de clusters et en stockant la valeur de la propriété <i>inertia_</i> à chaque fois. La metrique inertia est la WCSS (within-cluster sum of squares), c'est-a-dire, la somme carre des des distances des points à l'intérieur de chaque partition.

\begin{equation}
WCSS=\sum_{i=1}^{k}\sum_{x_j\in C_i}\|x_j-\mu_i\|^2
\end{equation}


```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

On analyse maintenant les differents solutions si on change les centroides initiales.

```python
#kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
wcss=[]
kmeans = KMeans(n_clusters=4,init='k-means++')
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=pred_y, s=50, cmap='copper')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='#DF88B7',alpha=0.7)
plt.show()
wcss.append(kmeans.inertia_)
```

```python

init_centers=np.array([[0,0],[2,0],[3,0],[3,1]])
kmeans = KMeans(n_clusters=4,init=init_centers)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=pred_y, s=50, cmap='copper')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='#DF88B7',alpha=0.7)
plt.show()
wcss.append(kmeans.inertia_)
```

```python
kmeans = KMeans(n_clusters=4,init='random')
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=pred_y, s=50, cmap='copper')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='#DF88B7',alpha=0.7)
plt.show()
wcss.append(kmeans.inertia_)
```

```python
init_centers=np.array([[-2,2],[-1,8],[1,4],[2,1]])
kmeans = KMeans(n_clusters=4,init=init_centers)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=pred_y, s=50, cmap='copper')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='#DF88B7',alpha=0.7)
plt.show()
wcss.append(kmeans.inertia_)
```

```python
centroides=np.array(['k-means++','[[0,0],[2,0],[3,0],[3,1]]','random','[-2,2],[-1,8],[1,4],[2,1]]'])
plt.plot(range(1, 5), wcss)
plt.xlabel('Choix de centroides')
plt.ylabel('WCSS: within-cluster sum of squares')
plt.xticks(range(1, 5),centroides)
plt.scatter([1,3,4], [wcss[0],wcss[2],wcss[3]], s=200, c='#DF88B7',alpha=0.7)
plt.show()

print(wcss)
```
